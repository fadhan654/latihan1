```python
import cv2
from google.colab.patches import cv2_imshow

# Baca gambar
image = cv2.imread("images.jpg")  # Sesuaikan nama file dengan file yang telah Anda unggah

if image is not None:  # Cek apakah gambar berhasil dimuat
    # 1. Menampilkan gambar asli
    print("Gambar asli:")
    cv2_imshow(image)

    # 2. Mengonversi gambar ke grayscale dan menampilkannya
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    print("Gambar dalam grayscale:")
    cv2_imshow(gray_image)

    # 3. Mengonversi gambar ke RGB dan menampilkannya
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    print("Gambar dalam model RGB:")
    cv2_imshow(rgb_image)

    # 4. Mengonversi gambar ke HSV dan menampilkannya
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    print("Gambar dalam model HSV:")
    cv2_imshow(hsv_image)

    # 5. Menampilkan komponen warna secara terpisah (R, G, B)
    print("Komponen merah (R):")
    red_component = image.copy()
    red_component[:, :, 0] = 0  # Hapus komponen hijau
    red_component[:, :, 1] = 0  # Hapus komponen biru
    cv2_imshow(red_component)

    print("Komponen hijau (G):")
    green_component = image.copy()
    green_component[:, :, 0] = 0  # Hapus komponen merah
    green_component[:, :, 2] = 0  # Hapus komponen biru
    cv2_imshow(green_component)

    print("Komponen biru (B):")
    blue_component = image.copy()
    blue_component[:, :, 1] = 0  # Hapus komponen merah
    blue_component[:, :, 2] = 0  # Hapus komponen hijau
    cv2_imshow(blue_component)

    # 6. Menerapkan deteksi tepi (Canny) dan menampilkannya
    edges = cv2.Canny(gray_image, 100, 200)
    print("Deteksi tepi menggunakan Canny:")
    cv2_imshow(edges)

    # 7. Melakukan rotasi gambar sebesar 45 derajat
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
    rotated_image = cv2.warpAffine(image, matrix, (w, h))
    print("Gambar yang dirotasi 45 derajat:")
    cv2_imshow(rotated_image)

else:
    print("Gambar gagal dimuat.")
```
